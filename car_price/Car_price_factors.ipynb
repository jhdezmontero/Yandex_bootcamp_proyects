## 2.1 Type of traction **`is_4wd`** 

`is_4wd` indicates whether a car is a four-wheel drive (i.e. 4wd) car. The database indicates that the car is 4wd by assigning a value of '1'. It is possible that cases with missing values represent cars without 4wd (i.e. 2wd).

Some car models are 4wd in their base version. However, other models have both 2wd and 4wd versions. To impute the values, let's see which models share the set of cars with 4wd (`cars_4wd`) and the set of cars with missing values (`cars_nawd`) using the numpy function **`intersecct1d`**.

```python
# Number of unique models
print('Total car models:',len(cars['model'].unique()),'\n')

# Models with 4wd 
cars_4wd = cars.loc[cars['is_4wd']==1, 'model'].unique()

# Models with missing values for 'is_4wd'
cars_nawd = cars.loc[cars['is_4wd'].isna(), 'model'].unique()

# Number  of models with 4wd
print('Models with 4wd:',len(cars_4wd),'\n')

# Number of models without specified traction
print('Models without specified traction:', len(cars_nawd),'\n')

# Models with 4wd in the set with missing values ('cars_nawd')
interseccion_wd = np.intersect1d(cars_4wd, cars_nawd, assume_unique=True, return_indices=False)

# Number of models that share cars_4wd and cars_nawd
print('Shared models between the two groups:',len(interseccion_wd))
```

It appears that all models in the `cars_4wd` dataset are found in the `cars_nawd` dataset with a total of 94 models. 

Due to the 100% intersection between the two datasets, it is unlikely that all models in our database are exclusively 4wd vehicles.

For now, we will attribute to the cases with missing values the value of zero (`0`) to indicate that the unit does not have 4wd.

Later we will convert the 1/0 values to True/False.

```python
# Replace NaN in 'is_4wd' with 0
cars['is_4wd'] = cars['is_4wd'].where(cars['is_4wd'] == 1, 0)

# Verify missing values
print('Casos con valores ausentes para "is_4wd":',cars['is_4wd'].isna().sum())
```

## 2.2 Number of cylinders **`cylinders`** 

There are 5260 cases with missing values for `cylinders`, because **the number of cylinders is a characteristic of the vehicle model that is usually constant**, we will explore which models do not have this information. Subsequently, we will impute the number of cylinders according to the model. 

```python
# Vehicle models with missing values in cylinders
cars_nacyl = cars.loc[cars['cylinders'].isna(), 'model'].unique()

print(len(cars_nacyl))
```

We have 100 different models that have no information. To impute missing values, let's **check if each car model has a unique number of cylinders**.

```python
# Group by car model and cylinders and obtain the frequency save in 'model_cyl'.
model_cyl = cars.pivot_table(index=['model','cylinders'], 
                             aggfunc='size').reset_index()

# Renname columns 'model_cyl'
model_cyl.columns = ['model', 'cylinders', 'frequency']

# Order values by leaving the number of cylinders with the highest frequency at the beginning of each model.
model_cyl = model_cyl.sort_values(by=['model','frequency'], ascending=[True,False])
model_cyl.head(10)
```

Our grouped table `model_cyl` shows per car model how many records there are for each `cylinder` level. In the first rows of the table, we can see that **there is no unique number of cylinders per model**, however when looking at the frequencies (column `frequency`) we can see that some `cylinder` levels have very low values (e.g. see model _cadillac escalade_). 

It is likely that the owner or the person in charge of registering the vehicle had no idea of the exact number of cylinders of the car model. To impute values we will take the `cylinders` level with the highest number of records and store these values in `good_cyl` to later impute `cylinders` according to the car model using the `join()` method.

The justification for making this change is that **the number of cylinders does not vary within a car model** and when there is variation the model name usually change to make a distinction in this characteristic.

```python
# Get the first record of the pivot table 'good_cyl'
good_cyl = model_cyl.groupby('model')['cylinders'].agg('first')

# Number of models in our 'good_cyl' series
print(len(good_cyl.index),'\n')

print(good_cyl,'\n')

# Impute values using join on 'model' method
# Imputed cylinders are stored in the variable 'cylinders_good'.
cars = cars.join(good_cyl, on='model', rsuffix='_good')
```

```python
# Number of missing values in 'cylinders_good'
cars['cylinders_good'].isna().sum()
```

```python
# Categories of before imputing cylinders
print(cars['cylinders'].unique())

# Categories after imputing cylinders
print(cars['cylinders_good'].unique())
```

We have now imputed the correct displacement to each case according to the car model and taking into account the highest number of registrations per number of cylinders. 

This makes sense because usually cars with odd numbers of cylinders (i.e. 3, 5) or larger than 8 cylinders (i.e. 10, 12) are not manufactured.

The imputed values are found in the **`cylinders_good`** column. 

## 2.4 Mileage **`odometer`**

We will estimate the missing values of `odometer` according to the distribution of the data to estimate whether to impute using the average or the median. We will aggregate the data by `model_year` and `condition` since these variables reflect car usage.

```python
# Adds the vehicle's average mileage per year and its condition
cars_odo = cars.pivot_table(index=['model_year','condition'], values='odometer', 
                            aggfunc=['mean','std','median','max','min','count']).reset_index().sort_values(by='model_year',ascending=False)
cars_odo.head()
```

Due to the wide range of values in the millage of some cars, we will **use the median values** as a more suitable centrality measure to impute missing values.

To impute the values we will create a dictionary where the keys is a tuple of `condition` and `model_year` and the value is the median of `odometer` for the combination of the tuple.

Finally, we will create a new column called `mileage` to store the values.

```python
# Create a dictionary where the tuple (condition_id, 'model_year') 
# is the key and the value is the median of odometer
odo_dict = cars.groupby(['condition','model_year'])['odometer'].median().to_dict()

# Function to obtain values from the dictionary 
# according to the combination of the tuple
def get_mpy(row):
    odometer = row['odometer']
    condition = row['condition']
    model = row['model_year']
    if np.isnan(odometer):
        return odo_dict.get((condition, model))
    else:
        return odometer
```

```python
# Create column 'mileage'
cars['mileage'] = cars.apply(get_mpy, axis=1)
```

```python
# Verify 'mileage' missing values
print(cars['mileage'].isna().sum())
```

After creating the `mileage` variable we still have 556 cases with missing values, probably because there is no combination of `condition` and `model_year` with data to get the median. Let's explore those cases.

```python
# Cases where there is no data for mileage 
cars.loc[cars['mileage'].isna()].head()
```

The first rows where there is still missing data show that **there is no data for `model_year` so in the absence of this value, it was not possible to impute the mileage**. Let's leave this imputation for now and deal with the missing values with another strategy later.

## 2.5 Model **`model_year`**

We will employ the same strategy to impute the model year from the model and condition of the car. Because the year is an integer type value, **we will use the median** and round to have non-fractional numbers.

We will store the imputed values in the `car_year` column without replacing the existing values in `model_year`.

```python
model_year_dict = cars.groupby(['model','condition'])['model_year'].median().to_dict()

# Function to obtain values from the dictionary according 
# to the combination of the tuple
def get_year(row):
    year = row['model_year']
    model = row['model']
    condition = row['condition']
    if np.isnan(year):
        return model_year_dict.get((model, condition))
    else:
        return year
```

```python
# Create 'car_year' column
cars['car_year'] = cars.apply(get_year, axis=1)

# Round 'car_year'
cars['car_year'] = cars['car_year'].apply(np.ceil)
```

```python
# Verify 'mileage' missing values
print(cars['car_year'].isna().sum())

# Very cases where 'cars_year' is NaN
cars.loc[cars['car_year'].isna()]
```

There are only 6 cases in which it was not possible to impute the model year, for the time being we will leave them as they are.

## 5.3 `transmission`

Since the transmission of a car can influence its price, being generally the automatic version models more expensive than the manual version, let's verify that this variable is correctly assigned.

**The assignment considers only the existence of two transmissions: automatic and manual**. Although there are other types of transmissions such as triptonic, these can be considered a variant of automatic transmission. The reduction of categories is done in order to simplify the analysis of the factors that influence the price of a vehicle.

```python
# Number of categories
cars['transmission'].value_counts()
```

We have 3 transmission categories `'automatic'`, `'manual'` and `'other'`. Although car models can be either automatic or manual, let's look at the frequency distribution of these categories by car model. 

We will select which transmission has the highest number of cases without taking into account the `'other'` category and assign for each model the preferred transmission.

```python
cars_trms = cars.pivot_table(index='model', columns='transmission', aggfunc='size').reset_index()
cars_trms['majority'] = cars_trms[['automatic','manual']].max(skipna=True, axis=1)
print(cars_trms['majority'].isna().sum())
cars_trms.head(10)
```

```python
# Function to get the 'dominant' transmission per model
def get_max_trms(row):
    automatic=row['automatic']
    manual=row['manual']
    other=row['other']
    majority=row['majority']
    
    if majority == automatic:
        return 'automatic'
    if majority == manual:
        return 'manual'
```

```python
cars_trms['preferred_trms'] = cars_trms.apply(get_max_trms, axis=1)
```

```python
print('Number of models with mostly automatic transmissions:',len(cars_trms.query('preferred_trms == "automatic"')))
print('Number of models with mostly manual transmission:',len(cars_trms.query('preferred_trms == "manual"')))
print('Number of models with mostly "other" transmission:',len(cars_trms.query('preferred_trms == "other"')))
cars_trms.head()
```

All models feature a preferred ``automatic'`` type transmission. No models have a majority with a ``manual`` preferred transmission. 

The manufacture and sale of vehicles is dominated by automatic models, so a higher frequency of models with automatic transmissions was expected. A vehicle model may have both automatic and manual versions while other models are only available in one version. For example, the _toyota prius_ model is a vehicle that is manufactured exclusively in automatic version.

Let's check what percentage of cases per car model are automatic. **If the percentage of automatic cars per model is greater than 95% of the total cases (adding only the ``automatic`` and ``manual`` cases) we will reclassify the model as automatic only**.

```python
# Obtain percentage of vehicles with automatic transmission without counting 'other'.
cars_trms['automatic_pct'] =  round(cars_trms['automatic'] / (cars_trms['automatic'] + cars_trms['manual']),2)
cars_trms.head()
```

```python
# Select models with a percentage of automatic versions > 0.95
print('Models with dominance >95% in automatic version:',len(cars_trms.query('automatic_pct > 0.95')))
print('Models without dominance >95% in automatic version:',len(cars_trms.query('automatic_pct < 0.95')))
print('Models with NA in "automatic_pct":',cars_trms['automatic_pct'].isna().sum())

manual_chg = cars_trms.loc[cars_trms['automatic_pct']>0.95, 'manual'].sum()
print(f'Number of cases to change (manual to automatic): {manual_chg}\nPorcentage of cases {manual_chg/len(cars):.2%}')
```

```python
# Verification of null values in 'automatic_pct'.
cars_trms.loc[cars_trms['automatic_pct'].isna()]
```

```python
# Change from NaN a 1
cars_trms.loc[cars_trms['automatic_pct'].isna(), 'automatic_pct'] = 1

# Select models with a percentage of automatic versions > 0.95
print('Models with dominance >95% in automatic version:',len(cars_trms.query('automatic_pct > 0.95')))
print('Models without dominance >95% in automatic version:',len(cars_trms.query('automatic_pct < 0.95')))
print('Models with NA in "automatic_pct":',cars_trms['automatic_pct'].isna().sum())

manual_chg = cars_trms.loc[cars_trms['automatic_pct']>0.95, 'manual'].sum()
print(f'Numero de casos a cambiar (manual a automatic): {manual_chg}\nPorcentaje de casos {manual_chg/len(cars):.2%}')
```

There are 67 models in which more than 95% are in automatic transmission version while 29 models do not show this dominance. We will reclassify the 56 models with automatic version dominance as exclusively automatic models.

**The existence of models registered as manual version may be due to filling errors in the registration form**. We consider that a threshold of 95% is conservative to make this change and **only affects 475 cases which represents less than 1% of the data**. We will store the new classification in the **`transmission_good`** column.

```python
# List of models with automatic dominance
model_atm = cars_trms.loc[cars_trms['automatic_pct']>0.95, 'model'].to_list()

# Create column 'transmission_good'
cars['transmission_good'] = cars['transmission']

# Replace values
cars['transmission_good'] = cars['transmission_good'].where(~cars['model'].isin(model_atm), 'automatic')
```

```python
# Verify changes
print(cars.groupby('transmission')['model'].nunique(),'\n')
print(cars.groupby('transmission_good')['model'].nunique(),'\n')
print(cars.groupby('transmission_good')['model'].count())
```

The number of car models with manual version has been reduced from 89 to 33. We still have 27 models classified as 'other' with 606 cases, we will review those cases to see if it is possible to impute any category.

```python
# List of models with 'other' transmission majority
model_otr = cars.loc[cars['transmission_good']=='other', 'model'].unique()
print('Models categorized as "other":', cars.loc[cars['transmission_good']=='other', 'model'].nunique(),'\n')

# Filter 'model_otr' en cars_trms
print('Total cases:',cars_trms.query('model in @model_otr')['other'].sum())
cars_trms.query('model in @model_otr')
```

The cases that were kept as ``other`` are because they did not reach the 95% automatic transmission dominance threshold.

The transmission allocation strategy (`transmission_good`) for the cases classified as ``other`` will be to assign the transmission with the majority of cases. Since in all cases the preferred transmission is ``automatic`` we will assign these 606 (0.01% of the total data set) cases as automatic version.

```python
# Assign 'automatic' for the remaining cases
cars.loc[cars['transmission_good']=='other', 'transmission_good'] = 'automatic'

# Check counts by categories of 'transmission_good'.
print(cars.groupby('transmission_good')['model'].count())
```

We reduced the number of transmission categories from 3 to 2, keeping only ``automatic'`` and ``manual'``. The strategy was based on imputing values according to a criterion of >95% of cases to determine exclusive transmission of the model; and relative majority to determine cases categorized as ``other'`` that did not meet an exclusivity threshold. 

**The total number of cases affected by this imputation strategy is 1081 cases, which represents only 0.02% of the total cases. We believe that this imputation is appropriate since the transfer of the vehicle may affect the selling price of the vehicle**.

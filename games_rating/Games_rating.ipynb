## Handling missing values <a id='2.3' ></a>

In this section, missing values will be explored and the best way to impute them will be determined if possible. Let's first explore the number of cases with missing values per column. Later we will analyze the missing values column by column and decide what to do with the missing values.

```python
# Missing values per column
games.isna().sum()
```

### Missing values `name`

### Missing values `year`

We have 269 cases whose release year (`year`) is not specified. We are going to see the number of different video games without release year and the first rows of the data subset with missing values in this column.

```python
# Games without year
print('Titles without launching year: ', games.loc[games['year'].isna()]['name'].nunique())
```

We have **232 *different* video game titles with no release year**. Because a game is published on more than one platform, it is possible to impute the value of each game's release year by taking the value of rows with one platform missing the year value. To impute the year of launch we will follow the following strategy:

1. Determine the titles that do not have a release year
2. For each title, determine if there is only one release year or if it has more
3. Impute values for those titles with only a *unique* year of release.
4. Impute values for titles with more than one year of release.

```python
# Games without launching year
no_release = games.loc[games['year'].isna()]['name'].unique().tolist()

# Determine if each game has only one launching year
games_years = games.loc[games['name'].isin(no_release)].groupby('name')['year'].nunique().sort_values(ascending=False).reset_index()

# Number of games with one or more launching years
games_years.value_counts('year')
```

Scanning for titles with missing release year values stored in the **games_years** table shows that they exist:

* 133 titles that have no way of imputing their release year
* 87 titles have an exclusive year of release
* 10 titles have two years of release
* 2 titles have 3 different launch years

First we are going to impute the values for the **87 titles that have an exclusive year of release**.

```python
# Titles with only one lauching year
games_one_year = games_years[games_years['year']==1]['name'].tolist()

# Get year
year_games_one_year = games.loc[games['name'].isin(games_one_year)].pivot_table(index='name', values='year', aggfunc='max').reset_index()

# Impute values
games = games.merge(year_games_one_year, on='name', how='left', suffixes=('_',''))
games['year'] = games.pop('year_').fillna(games['year'])

# Number of cases with missing values for 'year'
games['year'].isna().sum()
```

The release year has been imputed for those titles with only one release year. This reduced the number of records with missing values in `year` from 269 to 158, bringing the number of imputed cases to 111.

Now we are going to analyze the 10 titles that have **two different release years**.

```python
# List of titles with two different years of release
games_two_year = games_years[games_years['year']==2]['name'].tolist()
games.loc[games['name'].isin(games_two_year)].sort_values(by=['name','year'], ascending=[True,True])
```

When observing the titles with two different years of release, it is decided to take the minimum value of the year of release to impute the values. This is because video games with a year in their title like `'College Hoops 2K6'`, `'PES 2009: Pro Evolution Soccer'`, `'Tomb Raider (2013)'`, are usually published the previous year or the same year. to the one indicated in the title. In addition, the years of release were corroborated in the Internet Game Data Base database available online [IGBD] (https://www.igdb.com/).

We are going to use the minimum value to impute values for the 10 cases with a missing year of release, except for the title `'Sonic the Hedgehog'` for PS3 published in 2006 and which is duplicated. So we are going to remove this case from the database using its index **4126**. Later we will review duplicate values in detail.

```python
# Eliminate duplicated case
games = games.drop(4126)

# Uodate videogames without launching year 
no_release = games.loc[games['year'].isna()]['name'].unique().tolist()

# Determine if there is only one launching year 
games_years = games.loc[games['name'].isin(no_release)].groupby('name')['year'].nunique().sort_values(ascending=False).reset_index()

games_two_year = games_years[games_years['year']==2]['name'].tolist()

# Impute mininum value for 'year' 
year_games_two_year = games.loc[games['name'].isin(games_two_year)].pivot_table(index='name', values='year', aggfunc='min').reset_index()

year_games_two_year
```

Now we are going to establish a single value for cases with two different years of release.

```python
# Impute values and set only one release year
games = games.merge(year_games_two_year, on='name', how='left', suffixes=('_',''))

# Drop repeated column
games['year'] = games.pop('year_').fillna(games['year'])
```

We have imputed a single value for cases with two different years of release. Now we will see the cases with **three different years of release**. There are only two titles with this discrepancy. As there is so little data, the value will be manually imputed by consulting the launch year in [IGDB](https://www.igdb.com/).

The titles with missing values are the following:

* `'Space Invaders'` for Atari 2600
* `'Wheel of Fortune'` for PS2

According to the IGDB, the years of launch of these titles for each platform are 1978 and 2003 respectively.

```python
# Cases with three launching years 
games_three_year = games_years[games_years['year']==3]['name'].tolist()
games.loc[games['name'].isin(games_three_year)].sort_values(by='name')
```

```python
# Space Invaders para Atari 2600
games.loc[(games['name']=='Space Invaders')&(games['platform']=='2600'),'year'] = 1978

# Wheel of Fortune para PS2
games.loc[(games['name']=='Wheel of Fortune')&(games['platform']=='PS2'),'year'] = 2003

# Valores ausentes en año de lanzamiento
print(games['year'].isna().sum())

# Titulos sin referencia
games_no_year = games_years[games_years['year']==0]['name'].tolist()
games.loc[games['name'].isin(games_no_year)]['name'].nunique()
```

We have imputed values for video games that have a second reference value for their release year, thus reducing the cases with missing values from 269 to 146. The 146 cases correspond to 133 titles. As there is no reference that allows the values to be massively imputed, it is decided to leave them as absent values.

### Missing values `rating`

Next we will impute the missing values in `rating`. Like the year of release of each video game, its classification is an invariable attribute, so we are going to impute missing values when there is a reference value (i.e. same title, different platform).

The imputation of the `rating` will not be only for the cases with absent values. As the same title is released on different platforms (i.e. consoles), it is expected that the rating of the title will not be different between different platforms. We are going to determine if this is so and for the cases in which a title has more than one rating, we are going to impute the value with the highest frequency (i.e. if a title has 4 platforms and in 3 of them the category ` is assigned 'E'`, all games will be assigned the category `'E'`).

```python
# Get the category with the highest frequency per game
# Pivot table to group by title and number of categories
name_rating = games.pivot_table(index=['name','rating'], aggfunc='size').reset_index()

# Rename columns pivot_table name_rating
name_rating.columns = ['name','rating','frequency']

# Sort values by name (alphabetical) and frequency (highest to lowest)
name_rating = name_rating.sort_values(by=['name','frequency'], ascending=[True, False])

# Select rating row with the largest number of cases
rating_good = name_rating.groupby('name')['rating'].agg('first').reset_index()

# Number of titles to unify
print('Titulos a unificar: ', rating_good['name'].nunique())

# Total titles in the database
print(games['name'].nunique())

rating_good.head(10)
```

We have a list of 6066 of 11559 titles in total for which a single rating category will be standardized.

```python
# Standardize rating categories
games = games.join(rating_good.set_index('name'), on='name', rsuffix='_good')

# Keep only the rating_good column and rename it to rating
games.drop('rating', axis=1, inplace=True)
games.rename(columns={'rating_good':'rating'}, inplace=True)

# Number of missing values in 'rating'
print('Missing values for rating: ', games['rating'].isna().sum())

# Number of titles without 'rating'
print('Titles without rating: ', games.loc[games['rating'].isna()]['name'].nunique())
```

The number of missing values in `rating` has been reduced from 6766 to 6327, so only 439 values were imputed according to the category with greater frequency. The number of **unrated video game titles is 5493**. Because there is no reference to impute the `rating` to these games, they will be assigned the `'TBD'` rating, which means 'to be determined'.

```python
# Assign TBD to unrated titles
games.loc[games['rating'].isna(), 'rating'] = 'TBD'

# Missing values in rating
games['rating'].isna().sum()
```

### Missing values `critic_score` y `user_score`

There are **8576 cases with missing values in `critic_score`** and **9123 cases with missing values in user_score**. However, since some questions in the project are directly related to these metrics and since the ratings of the same title can vary widely from one platform to another, **no missing values** will be imputed.

### Conclusions: handling missing values

Missing values were treated in 6 variables, the imputation of values was made using reference values thanks to the fact that some video games were published on more than one platform.

* Removed 2 cases that had no information in both `name` and `genre`.
* The year of launch was imputed to 123 cases out of 269 cases with missing values in `year` (45%).
* The `rating` was assigned to all the titles when they had reference values, those without a reference value were assigned the `'TBD'` category.
* Missing values were not imputed for `critic_score` or `user_score` because they are directly related to the project questions and because scores can vary from platform to platform.

The database could not be completely purged of missing values due to the lack of reference values for all titles. The occurrence of these missing values may be due to the lack of ratings from both users and critics, either because the game is too old or because it is a recent release. In any case, it is advisable to update the database either with complementary data from other information sources such as the IGDB.

To show this we are going to analyze and compare the lifetimes of **Nintendo** consoles (`'NES'`, `'SNES'`, `'N64'`, `'GC'`, `'Wii' ` and `'WiiU'`) which span from 1983 to 2016.

A table will be created with the data of the Nintendo consoles called **`nintendo_plat`** which will contain the years of appearance (`first`) and disappearance (`last`) of each model. The time it takes for a platform to appear and disappear will be calculated as follows:

* Time to appear (`time_appear`): for each console, subtract the year of appearance of the next platform model (see `next_console` column) and the year of appearance of the console (`first`). This returns the number of years since the appearance of one platform and another.

* Time to disappear (`time_disappear`): for each console the year of appearance of the predecessor model (`next_console`) will be subtracted from the last year of registration of the console (`last`). This returns the time a console remains current after the release of a newer model.

The average number of years of both the time of appearance and the time of disappearance of the platforms will be calculated. For this calculation, the WiiU console will not be taken into account since it is the last Nintendo console registered in our database.

```python
# Nintendo Consoles 
nintendo = ['NES','SNES','N64','GC','Wii','WiiU']

# Slice Nintendo consoles
nintendo_plat = platform_years.query('platform in @nintendo').reset_index(drop=True)

# Take first 5 values the first to calculate differences in time 
next_console = nintendo_plat.iloc[1:, 1]
last = pd.Series([2016])
next_console = pd.concat([next_console,last], ignore_index=True)

# Add column to create differences
nintendo_plat['next_console'] = next_console

# time_appear
nintendo_plat['time_appear'] =  nintendo_plat['next_console'] - nintendo_plat['first']

# time_disappear
nintendo_plat['time_disappear'] = nintendo_plat['last'] - nintendo_plat['next_console']

# Average number of years to appear
time_appear_mean = nintendo_plat.loc[nintendo_plat['time_appear']>4]['time_appear'].mean() # No se toma en cuenta WiiU
time_appear_std = round(nintendo_plat.loc[nintendo_plat['time_appear']>4]['time_appear'].std(),2) # No se toma en cuenta WiiU
print(f'Average years for a new console to appear {time_appear_mean} ± {time_appear_std} years\n')

# Average number of years to disappear
time_disappear_mean = nintendo_plat.loc[nintendo_plat['time_disappear']>0]['time_disappear'].mean() # No se toma en cuenta WiiU
time_disappear_std = round(nintendo_plat.loc[nintendo_plat['time_disappear']>0]['time_disappear'].std(),2) # No se toma en cuenta WiiU
print(f'Average years for a console to disappear {time_disappear_mean} ± {time_disappear_std} years\n')

# Ver datos para plataformas de nintendo.
nintendo_plat
```

For **Nintendo** consoles, we find that the time (mean ± standard deviation) to **appear** a new console is **5.8 ± 0.84 years**. Once a new console appears, it takes **3 ± 1.22 years for the old model to disappear**.

```python
# Function to order the boxes of the graph according to their median from highest to lowest
def boxplot_sorted(df, by, column, rot=0, order=False):
    # create a dictionary to make a dataframe from an iterable groupby
    # each group becomes a column in the new data frame
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    # find and order the medians in df2 from largest to smallest default False
    meds = df2.median().sort_values(ascending=order)
    # use the columns in the dataframe, sort by median value
    # return axes in such a way that changes can be made outside of the function
    return df2[meds.index].boxplot(rot=rot, return_type="axes")
```

```python
# Boxplot ordered by median global sales
axes = boxplot_sorted(games_selected, by=['genre'], column='log_gl_sales', rot=45)
plt.title('Ventas globales por genero')
plt.xlabel('Género')
plt.ylabel('Millones de dolares (log10)')
plt.show()
```

# Hypothesis Test About Video Game Ratings <a id='hipotesis'></a>

In this **fifth stage** we are going to do a hypothesis test to compare the user ratings (`user_score`) between different groups. Specifically we are going to compare:

1. [Average Xbox One *vs* PS4 ratings](#5.1).
2. [Average Grades Action *vs* Sports](#5.2).

The data to be used to test these hypotheses will be taken from the **`games_selected`** database, which contains information for 5 platforms from 2011 to 2016.

## Average ratings Xbox One *vs* PS4 <a id='5.1'></a>

Let's compare the average ratings given by users for games released for Xbox One (`'XOne'`) and `'PS4'`. These two platforms are the most recent and have great popularity. Usually both platforms share several titles and the gaming experience is not expected to change from one platform to another. So we are going to test the following alternative and null hypothesis.

**$H_1$**: Ratings between XOne and PS4 users are *not* different.

**$H_0$**: Ratings between XOne and PS4 users are *yes* different.

To test these hypotheses we are going to select only the games that have a rating and that are shared by both platforms. The filtered data will be stored in the **'`ps4_xone_rates`** table. The means of the two populations will be obtained using the two-tailed t-test.

```python
# Data filter

# PS4 games with no missing values in 'user_score'
scores_ps4 = games_selected.loc[(games_selected['platform']=='PS4')&(~games_selected['user_score'].isna())]

# Games for XOne with no missing values in 'user_score'
scores_xone = games_selected.loc[(games_selected['platform']=='XOne')&(~games_selected['user_score'].isna())]

# Inner games shared between platforms
ps4_xone_rates = scores_ps4[['name', 'user_score']].merge(scores_xone[['name','user_score']], 
                                                          on='name', how='inner', suffixes=['_ps4', '_xone'])

# DataFrame Size
ps4_xone_rates.shape
```

We have 154 titles shared between both platforms. We are now going to buy the average value of the ratings received between platforms.

```python
# homoscedasticity
st.levene(ps4_xone_rates['user_score_ps4'], ps4_xone_rates['user_score_xone'])
```

```python
# Hypothesis test

# Significance level
alpha = 0.05

# Average user_score
print('Average user_score for PS4: ', ps4_xone_rates['user_score_ps4'].mean())
print('Average user_score for XOne: ', ps4_xone_rates['user_score_xone'].mean(),'\n')

# Variance user_score
print('Variance of user_score for PS4: ', np.var(ps4_xone_rates['user_score_ps4']))
print('Variance of user_score for XOne: ', np.var(ps4_xone_rates['user_score_xone']),'\n')

# Standard deviation
print('Standard deviation user_score for PS4: ', np.std(ps4_xone_rates['user_score_ps4']))
print('Standard deviation de user_score for XOne: ', np.std(ps4_xone_rates['user_score_xone']),'\n')

# t-test unilateral 
results = st.ttest_ind(ps4_xone_rates['user_score_ps4'], ps4_xone_rates['user_score_xone'])

# P-value
print('p-value: ', results.pvalue)

if results.pvalue > alpha:
    print("Reject null hypothesis")
else:
    print("Don't reject null hypothesis")
```

According to our hypothesis test, we reject the null hypothesis that stipulates that the user ratings of `'PS4'` and `'XOne'` are indeed different, so we have evidence in favor of our alternative hypothesis and we can say that **ratings for titles do not vary between platforms**.

## Average ratings for Action games *vs* Sports <a id='5.2'></a>

We are going to compare the rating that action and sports games receive across 5 platforms. Since player ratings can vary depending on the game genre and the popularity of the action games shown above, we are going to test the following alternative and null hypothesis.

**$H_1$**: The rating of action and sports games is different.

**$H_0$**: The rating of action and sports games is the same.

To test these hypotheses we are going to select from the **`games_selected`** table only the games of the `'Action'` and `'Sports'` genres that have user ratings. The filtered data will be stored in the **`action_rates`** and **`sport_rates`** tables. The means of the two populations will be obtained using the two-tailed t-test.

```python
# Data selection
action_rates = games_selected.loc[(games_selected['genre']=='Action') & (~games_selected['user_score'].isna())]
sport_rates = games_selected.loc[(games_selected['genre']=='Sports') & (~games_selected['user_score'].isna())]

# Cases per game type
print('Action:',action_rates.shape[0],'\n','Sports:', sport_rates.shape[0])
```

```python
# Homocedasticity
st.levene(action_rates['user_score'], sport_rates['user_score'])
```

```python
# Hypothesis test

# Significance level
alpha = 0.05

# Average user_score 
print('Average user_score for Action games: ', action_rates['user_score'].mean())
print('Average user_score for Sports games: ', sport_rates['user_score'].mean(),'\n')

# Variance
print('Variance user_score for Action games: ', np.var(action_rates['user_score']))
print('Variance user_score for Sports games:  ', np.var(sport_rates['user_score']),'\n')

# Standard deviation
print('Standard deviation usuer_score for Action games: ', np.std(action_rates['user_score']))
print('Standard deviation usuer_score for Sports games: ', np.std(sport_rates['user_score']),'\n')

# t-test unilateral 
results = st.ttest_ind(action_rates['user_score'], sport_rates['user_score'], equal_var=False)

# P-value
print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("Reject null hypothesis")
else:
    print("Don't reject null hypothesis")
```

The rating for action games is 6.76 ± 1.29 (mean ± standard deviation) and sports games is 5.57 ± 1.67. The p value was less than 0.5, so we can say that **the scores between both genders are significantly different**.
